**Covid-19 stats for various countires.**

```python
path2 = 'https://gist.githubusercontent.com/aakashns/b2a968a6cfd9fbbb0ff3d6bd0f26262b/raw/b115ed1dfa17f10fc88bf966236cd4d9032f1df8/covid-countries-data.csv'

covid_df1 = pd.read_csv(path2)
covid_df1.head()
```

**Q8: Count the number of countries for which the `total_tests` data is missing.**



```python
covid_df1.total_tests.isnull().sum()
```



**Q9: Merge `countries_df` with `covid_data_df` on the `location` column.**

*Hint: Use the `.merge` method on `countries_df`.

```python
combined_df = covid_df1.merge(countries_df, on='location')
combined_df.head()
```

**Q10: Add columns `tests_per_million`, `cases_per_million` and `deaths_per_million` into `combined_df`.**

```python
combined_df['tests_per_million'] = combined_df['total_tests'] * 1e6 / combined_df['population']
combined_df['cases_per_million'] = combined_df['total_cases'] * 1e6 / combined_df['population']
combined_df['deaths_per_million'] = combined_df['total_deaths'] * 1e6 / combined_df['population']
combined_df.tail()
```

**Q11: Create a dataframe with 10 countires that have highest number of tests per million people.**

```python
tpm = combined_df.nlargest(10, 'tests_per_million')
tpm[['location', 'tests_per_million']]
```

**Q12: Create a dataframe with 10 countires that have highest number of positive cases per million people.**

```python
cpm = combined_df.nlargest(10, 'cases_per_million')
cpm[['location', 'cases_per_million']]
```

**Q13: Create a dataframe with 10 countires that have highest number of deaths cases per million people?**

```python
dpm = combined_df.nlargest(10, 'deaths_per_million')
dpm[['location', 'deaths_per_million']]
```

**(Optional) Q: Count number of countries that feature in both the lists of "highest number of tests per million" and "highest number of cases per million".**

```python
resData = cpm.merge(tpm, how = 'inner' )
num_1 = resData.location.count()
print('There are {} countries that feature in both the lists. '.format(num_1))
```

**(Optional) Q: Count number of countries that feature in both the lists "20 countries with lowest GDP per capita" and "20 countries with the lowest number of hospital beds per thousand population". Only consider countries with a population higher than 10 million while creating the list.**

```python
# Countries having population higher than ten million
countries_df_10 = countries_df[countries_df['population'] > 10e6]
countries_df_10.head()
```

```python
# Twenty countries from above dataframe having lowest gdp capital
gpc_low = countries_df_10.sort_values('gdp_per_capita').head(20)

gpc_low
```

```python
# 20 countries with lowest number hospital bed per thousand population from countries_df_10
hbpt_low = countries_df_10.sort_values('hospital_beds_per_thousand').head(20)
hbpt_low
```

```python
com_countries = gpc_low.merge(hbpt_low, how='inner')
cu = com_countries.location.count()
print('There are {} countries in both list.'.format(cu))
```
